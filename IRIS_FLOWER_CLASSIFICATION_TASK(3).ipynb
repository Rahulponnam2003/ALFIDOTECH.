# **Iris flower classififcation**

The iris classification file contains iris specimens divided into three types:

1. Iris setosa
2. Iris versicolor
3. Iris virginica

This file is divided into three groups, there are 50 in each category representing different types of iris plants times that It is worth noting that one of the groups can be separated from the other two groups using a border line, while the other two groups are not linearly separated from each other.

The dataset contains the following attributes for each iris model:

1. Sepal length (in centimeters)
2. Sepal width (in centimeters)
3. Petal length (in centimeters)
4. Petal Centimeters Width in
5. Class


This item is used to measure iris flowers. This information is widely accepted and used as a reference point for tracking task classification in machine learning. The main purpose is to classify iris flowers according to their types according to these characteristics.

# **Model Training**

```python
x = df.drop(columns=['Species'])
y = df['Species']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

# **MODEL 1**

```python
# Logistic Regression Model
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(x_train, y_train)
accuracy_logreg = model1.score(x_test, y_test)
print("Accuracy (Logistic Regression): ", accuracy_logreg)
```

# **MODEL 2**

```python
# K-nearest neighbours Model(KNN)
from sklearn.neighbors import KNeighborsClassifier
model2 = KNeighborsClassifier()
model2.fit(x_train, y_train)
accuracy_knn = model2.score(x_test, y_test)
print("Accuracy (KNN): ", accuracy_knn)
```

# **MODEL 3**

```python
# Decision Tree Model
from sklearn.tree import DecisionTreeClassifier
model3 = DecisionTreeClassifier()
model3.fit(x_train, y_train)
accuracy_decision_tree = model3.score(x_test, y_test)
print("Accuracy (Decision Tree): ", accuracy_decision_tree)
```

# **PROJECT REPORT** MODEL COMPARSION VISULAISATION

```python
# Model Comparison - Visualization (COMPARISON OF 3 MODELS)
models = ['Logistic Regression', 'KNN', 'Decision Tree']
accuracies = [accuracy_logreg, accuracy_knn, accuracy_decision_tree]

plt.bar(models, accuracies, color=['blue', 'green', 'orange'])
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Model Comparison - Accuracy")
plt.ylim([0, 1])
plt.show()
```
